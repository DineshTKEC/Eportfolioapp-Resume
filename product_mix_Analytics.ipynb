```r
# ---------------------------------------
# Step 1: Install & Load Required Package
# ---------------------------------------
install.packages("ISLR")     # Only once
library(ISLR)

# ------------------------
# Step 2: Load the Dataset
# ------------------------
data("Carseats")
head(Carseats)

# ------------------------------------------
# Step 3: Summary of Data (Exploratory Step)
# ------------------------------------------
summary(Carseats)
str(Carseats)

# --------------------------------------------
# Step 4: Linear Regression - Sales Prediction
# --------------------------------------------
# Using variables related to product mix & marketing
model <- lm(Sales ~ Price + Advertising + ShelveLoc + Age + Income + Urban + US, data = Carseats)

# --------------------------
# Step 5: Model Summary
# --------------------------
summary(model)

# -----------------------------------------------
# Step 6: Visualize Relationships (Optional Step)
# -----------------------------------------------
par(mfrow = c(2, 2))   # Plot layout
plot(model)

# --------------------------------------------------------
# Step 7: Key Insights (based on Coefficients and p-values)
# --------------------------------------------------------
cat("\n--- Key Insights ---\n")
cat("1. Price: Negative impact on sales — Lower price, higher sales.\n")
cat("2. Advertising: Positive coefficient — More ad spend increases sales.\n")
cat("3. ShelveLoc: Strong influence — Better shelf location drives higher sales.\n")
cat("4. Urban/US: Demographic variables — Consider for regional targeting.\n")

# ----------------------------------------------------------
# Step 8: Business Decision Support (Amul Product Strategy)
# ----------------------------------------------------------
cat("\n--- Strategic Applications for Amul ---\n")
cat("✅ Invest more in Advertising for new product launches.\n")
cat("✅ Negotiate for prime shelf space in retail outlets.\n")
cat("✅ Use price promotions tactically to boost lower-selling variants.\n")
cat("✅ Target urban vs. rural markets differently based on product line.\n")
```
